### Life Cycle of a Star

All stars form in nebulae, which are huge clouds of gas and dust. Though they shine for many thousands, and even millions of years, stars do not last for ever. The changes that occur in a star over time and the final stage of its life depend on a star's size. 
A smaller star, like the Sun, will gradually cool down and stop glowing. During these changes it will go through the planetary nebula phase, and white dwarf phase. After many thousands of millions of years it will stop glowing and become a black dwarf.

A massive star experiences a much more energetic and violent end. It explodes as a supernova. This scatters materials from inside the star across space. This material can collect in nebulae and form the next generation of stars. After the dust clears, a very dense neutron star is left behind. These spin rapidly and can give off streams of radiation, known as pulsars.

If the star is especially massive, when it explodes it forms a black hole.

### What is a Star

A star is a celestial body made of plasma, more precisely a spheroid, that shines with its own light. It generates energy from within its core due to thermonuclear fusion.

Stars appear, as well as individually, also in binary systems or multiple systems, bound together by the gravity force. A good number of stars live in clusters (opened or globular), in turn grouped, together with lone stars and clouds of dust and gas, in bigger clusters called galaxies. Stars can also have one or more planets.

Stars are objects that have considerable mass, among 0.08 and 150-200 solar masses (Mo). Stars with a mass below 0.08 Mo are called brown dwarfs, objects half way between a star and a planet, that do not produce energy through nuclear fusion, while stars with a mass over 200 Mo seem to not exist, due to the Eddington's limit. Dimensions vary too ranging between few km of degenerated stars and billions of km of supergiants and hypergiants, and luminosity among 10^-4^ and 106-107 Lo.

### Properties and star types related to the dataset variables

 #### Temperature(k) or Star color
 
 In astronomy, the variables temperature and color are closely connected, that's why they are explained together.

The many colors of the stars are determined by the most part of the wavelength of the visible radiation that they emit: grerater the wavelength, more reddish is the color. In turn, the wavelength of the radiation emitted from the star depends on the superficial (photosphere) temperature of the star. The photosphere has a color that depends on its temperature: if it's very hot, the star will appear white, if it's cold will appear yellow-orange, if it's even colder it will be red.

Everything can be summarise by the Wien's law: objects of different temperature emit spectra that peak at different wavelengths. Hotter objects emit most of their radiation at shorter wavelengths; hence they will appear to be bluer. Cooler objects emit most of their radiation at longer wavelengths; hence they will appear to be redder.

The temperature of the star is measured in Kelvin (K).

### Luminosity (L/Lo)

In astronomy, luminosity is the amount of electromagnetic energy emitted by a star per unit of time, that is its power. It is measured in watt, erg/sec or in solar luminosity.

Luminosity of a star depends on both its temperature and its superficial area and it's a base quantity in astronomy and astrophysics. Most of what we know about celestial bodies it has been learned analyzing their light. The reason is that physical process that occur into the stars are, for the most part, recorded and transmitted by the light.

### Radius (R/Ro)

Radius of stars is usually expressed relative to that of the Sun, that is equal to 6.9551x10^8^m. Bigger radii, like that of supergiants or hypergiants or semi-major axis of a binary system, are often expressed in astronomical unit (A.U.), that is equivvalent to the distance between the Sun and the Earth (about 150 millions km).

### Absolute magnitude

In astronomy, the absolute magnitude (M) is the apparent magnitude that an object would have if placed at 10 parsec (32.6 light years) from the observer. The most used absolute magnitude is the visual absolute magnitude Mv measured in the visible spectrum.

### Spectral Class

In astronomy, stellar classification is the classification of stars based on their spectral characteristics, that is, their temperature, since the two quantities are tightly connected.

Most stars are currently classified under the Morgan-Keenan (MK) system using the letters O, B, A, F, G, K, and M, a sequence from the hottest (O type) to the coolest (M type). It is customary to describe O class stars as "blue", B class as "blue white", A class as "white", F class as "yellow white", G class as "yellow", K class as "orange" and M class as "red".

### Star Type

Brown Dwarf: they are hybrid celestial bodies, half way between a star and a planet. They have a mass too small to generate the necessary temperature to start the nuclear reactions that make the stars bright. The minimal amount of mass to become a star is about 80 Jupiter masses. Brown dwarfs cannot be considered planets either. In fact, they are too massive, but like planets, they don't bright of their own light.

Red Dwarf: it's a little and relatively cold star (T ≤ 3500 K), M spectral class (photspheric color, on average, yellow - orange). It is the most common stellar typology in the universe.

White Dwarf: it's a small star with a very low luminosity and a color tending to white. Despite being small, comparable to those of the Earth, the mass of this star is like or a bit more of that of the Sun;it is therefore a very dense object, with a very high density and surface gravity.White dwarfs possess, in the time of their formation, a very high temperature that gradually decrease according to the thermal exchanges with the outer space.

Main Sequence: it is the part of the curve, inside the Hertzsprung-Russell diagram, that is located in the lower center. When a star is located in any point of this curve, it means that it is in the longest and stable part of its life, ranging from its stabilization after birth to the next stage called "T Tauri" (characterized by strong stellar wind that dispel the parent nebula), and lastly to the pulsating stage that declare the unbearable mass pressure of the star due to the incoming dry up of the hydrogen inside the stellar nucleus, with the "stop" of fusion reactions of the hydrogen into elium inside the nucleus.

Supergiant: in astronomy, supergiants are stars that are classified as I into the Yerkes spectral classification. Stars belonging to this class type show very thin spectral lines compared to those of the main sequence. The thinned lines are indication that the atmposphere of these stars are very rarefied. Most part of the supergiants are massive stars that in the final stage of their life, considerably increase their radius. This phenomenon has two implications, reduce the density of the star thinning the spectral lines and, hugely increasing the radiant surface, raises the luminosity of the star. Hence these supergiants fill the upper region of the H-R diagram, reserved to the brighter stars having an absolute magnitude between -5 and -12.

Hypergiant: it's a massive star (brightness class 0), bigger than a supergiant, having a mass that can go up to 100 times that of the Sun. They are connsidered the brightest existing stars and their temperature range among 3500K and 35000K. The length of their life is estimated in about two millions years, at the end of that they explode in brilliant supernovae or even hypernovae. It is theorized that a hypergiant, after ther explosion, would leave behind an extremely dense and massive black hole.

```python
y=df[['Temperature','L','R','A_M']]
```

```python
color_label={value: key for key, value in enumerate(df['Color'].unique())}
spectral_class_label={value: key for key, value in enumerate(df['Spectral_Class'].unique())}
df['Spectral_Class']=df['Spectral_Class'].map(spectral_class_label)
df['Color']=df['Color'].map(color_label)
```

```python
df['Type'].unique()
```

```python
x=df[['Temperature','L','R','A_M']].values
y=df[['Type']].values
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation Matrix',fontsize=20)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('xtrain shape:',x_train.shape)
print('xtest shape:',x_test.shape)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

### Model 1: Logistic Regression

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
pred=lg.predict(x_test)
```

```python
y_test=y_test.reshape(48,)
print(classification_report(y_test,pred))
```

```python
print(accuracy_score(y_test,pred))
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(confusion_matrix(y_test,pred),annot=True)
```
